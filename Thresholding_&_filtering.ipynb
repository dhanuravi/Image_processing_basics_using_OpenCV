#simple thresholding

```
import cv2
img = cv2.imread('lion.jpg')  
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
ret, thresh1 = cv2.threshold(img, 135, 255, cv2.THRESH_BINARY) 
plt.subplot(2,2,3),plt.imshow(img,'gray')
plt.title('Original image')
plt.xticks([]),plt.yticks([])
plt.subplot(2,2,4),plt.imshow(thresh1,'gray')
plt.title('Simple threshold - Binary')
plt.xticks([]),plt.yticks([])
```

#Smoothening

```
avg_blur = cv2.blur(img,(6,6))
plt.subplot(2,2,1),plt.imshow(img,'gray')
plt.title('Original Image')
plt.xticks([]),plt.yticks([])
plt.subplot(2,2,2),plt.imshow(avg_blur,'gray')
plt.title('Smoothened image')
plt.xticks([]),plt.yticks([])
```

Filtering

```
bf = cv2.imread('butter (3).jpg')
bf=cv2.resize(bf,(250,250))
Median_filter = cv2.medianBlur(bf,3)
gausion_filter = cv2.GaussianBlur(bf,(5,5),10)
bilateral_filter = cv2.bilateralFilter(bf,150,90,90)
plt.subplot(2,2,1),plt.imshow(bf,'gray')
plt.title('Original Image')
plt.xticks([]),plt.yticks([])
plt.subplot(2,2,2),plt.imshow(Median_filter,'gray')
plt.title('Median_filter')
plt.xticks([]),plt.yticks([])
plt.subplot(2,2,3),plt.imshow(gausion_filter,'gray')
plt.title('gausion_filter')
plt.xticks([]),plt.yticks([])
plt.subplot(2,2,4),plt.imshow(bilateral_filter,'gray')
plt.title('bilateral_filter')
plt.xticks([]),plt.yticks([])
```
