```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

Canny edge detection

```
img = cv2.imread('butter.jpg')
edges = cv2.Canny(img,3,5)
edges2 = cv2.Canny(img,50,60)
edges1 = cv2.Canny(img,100,200)
plt.subplot(2,2,1),plt.imshow(img,cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(edges,cmap='gray')
plt.title('Edge Image - 3,5'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(edges2,cmap='gray')
plt.title('Edge Image -50,60'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(edges1,cmap='gray')
plt.title('Edge Image  100,200'), plt.xticks([]), plt.yticks([])
plt.show()
```

SOBEL  edge detection

```
# converting to gray scale
#gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img1 = cv2.imread('butter.jpg',1)
# remove noise
img = cv2.GaussianBlur(img1,(3,3),0)

# convolute with proper kernels
#laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)  # x
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)  # y
sobelx1 = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)  # x1
sobely1 = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)  # y

plt.subplot(2,2,1),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel x'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(sobelx1,cmap = 'gray')
plt.title('Sobel x1'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel y'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely1,cmap = 'gray')
plt.title('Sobel Y1'), plt.xticks([]), plt.yticks([])

plt.show()
```

 LAPLACIAN Edge detection

```
# converting to gray scale
#gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img1 = cv2.imread('butter.jpg',0)
# remove noise
img = cv2.GaussianBlur(img1,(3,3),0)

# convolute with proper kernels
laplacian1 = cv2.Laplacian(img,cv2.CV_8U)
laplacian2 = cv2.Laplacian(img,cv2.CV_16S)
laplacian3 = cv2.Laplacian(img,cv2.CV_64F)


plt.subplot(2,2,1),plt.imshow(img1,cmap = 'gray')
plt.title('Orginal Image'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian1,cmap = 'gray')
plt.title('laplacian1'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(laplacian2,cmap = 'gray')
plt.title('laplacian2'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(laplacian3,cmap = 'gray')
plt.title('laplacian3'), plt.xticks([]), plt.yticks([])

plt.show()
```
