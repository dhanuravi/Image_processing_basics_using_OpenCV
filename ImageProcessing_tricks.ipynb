```
img1 = cv2.imread("lion.jpg")
img2 = cv2.imread('aloevera.jpg')
```

Guassian Blur

```
img1_g = cv2.GaussianBlur(img1,(5,5),10)
cv2_imshow(img1_g)
```

ImageBlend

```
img11 = cv2.resize(img1,(300,300))
img21 = cv2.resize(img2,(300,300))
blend = cv2.addWeighted(img11,0.5,img21,0.5,1)
#himg = hstack()
cv2_imshow(blend)
```
